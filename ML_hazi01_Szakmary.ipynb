##Zaj kép generálása

```python
zajkep = np.random.randint(0,255, size=hegy.shape)
plt.imshow(zajkep)
```

##A zaj kép RGB csatornákkal

```python
zajkep_red = zajkep.copy()
zajkep_green = zajkep.copy()

zajkep_red[:,:,(1,2)] = 0
plt.imshow(zajkep_red)
```

```python
zajkep_green[:,:,(0,2)] = 0
plt.imshow(zajkep_green)
```

##Képek megjelenítése egyszerre

```python
fig, axs = plt.subplots(2, 2)
axs[0,0].imshow(labda)
axs[0,1].imshow(hegy)
axs[1,0].imshow(hal)
axs[1,1].imshow(zajkep)
```

##Színcsatornák átlagos értékeinek kiszámolása

```python
rgb = ["piros", "zöld", "kék"]
kepek = [labda, hegy, hal, zajkep]
kepnev=["Labda", "Hegy", "Hal", "Zajkép"]

for i in range(4):
  print(f"\nA színcsatornák átlagos értékei a {kepnev[i]} képen:")
  for j in range(3):
    print(f"{rgb[j]} :{kepek[i][:, :, j].mean()}")
```
